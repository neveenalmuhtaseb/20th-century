```python
# Sentence tokenization

from nltk.tokenize import sent_tokenize
tokenized_sent = sent_tokenize(data)
print(tokenized_sent)
```

```python
# Word tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words) 
```

```python
dist_words.most_common(10)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words.plot(10, cumulative=False)
plt.show()
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

filtered_words = []  # creates an empty list
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
from nltk import FreqDist  # make sure this is included

# Create a new FreqDist for filtered_words
Dist_words_filter = FreqDist(filtered_words)
print(Dist_words_filter)
```

```python
# Frequency Distribution Plot
plt.figure(figsize=(8, 3))
Dist_words_filter.plot(10, cumulative=False)  
plt.show()
```

```python
# Substitute all punctuation marks with a space

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                        " ",        # Replace all non-letters with spaces
                        str(filtered_words))
```

```python
# Word tokenization
Tokenized_word_2 = word_tokenize(sans_punc)
print(Tokenized_word_2) 
```

```python
# Create a new FreqDist

dist_words_filter_2 = FreqDist(Tokenized_word_2) 
```

```python
# Frequency Distribution Plot
plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(10, cumulative=False)  
plt.show()
```

```python
new_stopwords = ["And", "Then", 'n', 't', 's', 'The','th']

filtered = []
for word in Tokenized_word_2:
     if word not in new_stopwords:
        filtered.append(word)
```

```python
%%time
text = TextBlob(str(filtered))
```

```python
tags_list = text.tags
```

```python
tags_list
```

```python
df_text = pd.DataFrame(tags_list)
df_text.columns = [ 'Words', "Word type"]
df_t=df_text.groupby('Word type').count().reset_index()
top20=df_t.nlargest(20, 'Words')
```

# Bar Plot for Nouns

```python
# Count frequencies for Nouns
nouns = [word for word, pos in tags_list if pos.startswith('NN')]  # Nouns

# Get the top 15 most common nouns
nouns_count = Counter(nouns).most_common(15)

# Plot for Nouns
plt.figure(figsize=(10, 6))
sns.barplot(x=[item[1] for item in nouns_count], y=[item[0] for item in nouns_count], palette='Blues_d')
plt.title("Top 15 Nouns", fontsize=16)
plt.xlabel("Frequency", fontsize=12)
plt.ylabel("Nouns", fontsize=12)
plt.show()
```

# Bar Plot for Verbs

```python
# Count frequencies for Verbs
verbs = [word for word, pos in tags_list if pos.startswith('VB')]  # Verbs
verbs_count = Counter(verbs).most_common(15)

# Plot the top 15 Verbs
plt.figure(figsize=(10, 6))
sns.barplot(x=[item[1] for item in verbs_count], y=[item[0] for item in verbs_count], palette='Greens_d')
plt.title("Top 15 Verbs", fontsize=16)
plt.xlabel("Frequency", fontsize=12)
plt.ylabel("Verbs", fontsize=12)
plt.show()
```

# Bar Plot for Adjectives

```python
# Count frequencies for Adjectives
adjectives = [word for word, pos in tags_list if pos.startswith('JJ')]  # Adjectives
adjectives_count = Counter(adjectives).most_common(15)

# Plot the top 15 Adjectives
plt.figure(figsize=(10, 6))
sns.barplot(x=[item[1] for item in adjectives_count], y=[item[0] for item in adjectives_count], palette='Purples_d')
plt.title("Top 15 Adjectives", fontsize=16)
plt.xlabel("Frequency", fontsize=12)
plt.ylabel("Adjectives", fontsize=12)
plt.show()
```

# Create a dataframe with the countries and the number of times they’re mentioned 

```python
import pycountry
import re
from collections import Counter
import pandas as pd

# Prepare a case-insensitive regex pattern for all country names
country_names = [re.escape(country.name) for country in pycountry.countries]
pattern = re.compile(r'\b(' + '|'.join(country_names) + r')\b', re.IGNORECASE)

# Find all matches (case-insensitive)
matches = pattern.findall(listToStr)

# Normalize names for counting (e.g., "united states" → "United States")
normalized_matches = [
    pycountry.countries.lookup(m.title()).name
    for m in matches if m.title() in [c.name for c in pycountry.countries]
]

# Count and build DataFrame
country_counts = Counter(normalized_matches)
df_countries = pd.DataFrame(country_counts.items(), columns=["Country", "Mentions"]).sort_values("Mentions", ascending=False)
df_countries
```

```python
# Plotting country mention frequencies

plt.figure(figsize=(12, 6))
plt.bar(df_countries['Country'], df_countries['Mentions'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Frequency of Country Mentions in the Text')
plt.xlabel('Country')
plt.ylabel('Mentions')
plt.tight_layout()
plt.show()
```

The frequency plot shows that **Germany**, **Japan**, and the **United States** are the most frequently mentioned countries in the 20th-century text. This aligns closely with their major roles in shaping global events during this era. 

- **Germany** and **Japan** were central to both World Wars, particularly World War II, where their military actions and eventual defeats had lasting global consequences.
- The **United States** emerged as a dominant global power in the 20th century—playing key roles in both World Wars, the Cold War, and international diplomacy.

Other countries like **France**, **Italy**, and **Poland** also had significant involvement in these conflicts and subsequent geopolitical shifts. **China**, **India**, and **Pakistan** reflect the decolonization movements and the emergence of new nation-states post-World War II. Mentions of **Israel** and **Libya** likely relate to Middle Eastern conflicts and postcolonial tensions.
